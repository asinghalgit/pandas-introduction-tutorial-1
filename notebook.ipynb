```python
weather_df = pd.read_csv("weather.csv")
```

```python
weather_df.head()
```

```python
weather_melted = weather_df.melt(id_vars=["id","year","month","element"], var_name="day",value_name="temp")
```

```python
weather_melted.head()
```

```python
weather_melted = weather_melted.pivot_table(index=["id","year","month","day"],columns="element",values="temp")
```

```python
weather_melted = weather_melted.reset_index()
```

```python
weather_melted.head()
```
